```python
import pandas as pd

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data = load_digits()
```

```python
dir(data)
```

```python
print(data.DESCR)
```

```python
digit_data = data.data
digit_data
```

```python
digit_label = data.target
digit_label
```

```python
data.target_names
```

```python
x_train ,x_test ,y_train , y_test = train_test_split(digit_data ,digit_label, random_state=42, test_size =0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression,SGDClassifier
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
svm_model = svm.SVC()
sgd = SGDClassifier()
lr = LogisticRegression()

model =[dt,rf,svm_model,sgd,lr]
```

```python
str(dt)
```

```python
for i in model:
  i.fit(x_train,y_train)
  y_pred = i.predict(x_test)
  print(str(i))
  print(classification_report(y_test,y_pred))  # 숫자들이 다수 나오기에 직관성을 떨어지나 해석하기에는 용이하다.
  print(confusion_matrix(y_test,y_pred))       # 시각적으로 행렬으로 나와 직관성은 좋으나 해석하기에 상대적으로 난해하다. 
  
  print('----------')
```
