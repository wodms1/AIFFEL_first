## Activation Function

- 활성화 함수: 입력 신호를 출력신호로 변환, → 입력값의 활성화 정도, 출력값에 비선형성 부여
     perceptron→ step function


- Sigmoid:$\frac{1}{1+e^{-x}}$
    - 입력값이 크다면  큰값을 출력 & 0~1사이값을 출력

- ReLU function
    - 일반적으로 사용되는 활성화 함수

- Softmax function :$\frac{e_a}{\sum_{i=1}^{n}e_{a_i}}$
    - ML → SL → 분류 → 범주형 분류에 사용된다
    - 출력층의 활성화 함수로 활용된다. 출력층 노드 수는 target수 이다.
    - 각 노드에는 softmax를 걸친 target별 확률값이 나온다 → 0~1 사이값
    - 출력층 노드의 활률 총합 == 1
- fix_softmax function
    - a값이 커진다면 지수함수는 무한대로 발산한다
    - 따라서 지수함수의 성질을 이용하여 max값을 분모분자에 곱한다.

## 신경망 구현

- 3층 forward propagation 구현

```python
def identity_function(x): # 항등함수
  return x

def init_network():       # 매개변수 초기화 함수 -> W , b
  network = {}
  network['W1'] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
  network['b1'] = np.array([0.1,0.2,0.3])
  network['W2'] = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
  network['b2'] = np.array([0.1,0.2])
  network['W3'] = np.array([[0.1,0.3],[0.2,0.4]])
  network['b3'] = np.array([0.1,0.2])
  return network

def forward(network,k):  # 순전파 함수
  W1,W2,W3 = network['W1'],network['W2'],network['W3']
  b1,b2,b3 = network['b1'],network['b2'],network['b3']

  a1 = np.dot(x,W1)+b1
  z1 = sigmoid(a1)
  a2 = np.dot(z1,W2)+b2
  z2 = sigmoid(a2)
  a3 = np.dot(z2,W3)+b3
  y = identity_function(a3)
  return y
  
```

```python
network = init_network()
x = np.array([1.0,0.5])
y = forward(network,x)
print(y)
```
