# Perceptron
- perceptron? :프랑크 로젠블랏이 1957년에 고안한 알고리즘으로 인간의 뇌의 뉴런을 모방한 알고리즘, 다수의 신호를 입력받아 하나의 신호를 출력한다.


```python
import numpy as np

def perceptron_AND(x1,x2):
		w1,w2,theta = 0.5,0.5,0.7
		if x1*w1 + x2*w2 > theta:
			return 1
		if x1*w1 + x2*w2 < theta:
			return 0

def perceptron_AND_(x1,x2):                    # 임계값이 편향으로 치환
		x = np.array([x1,x2])
		w = np.array([0.5,0.5])               # 입력신호 중요도 조절
		bias = -0.7                           # 뉴런 활성화 조절
		tmp = x*w + bias
		if tmp > 0:
			return 1
		else:
			return 0

def perceptron_NAND(x1,x2):
		x = np.array([x1,x2])
		w = np.array([-0.5,-0.5])
		b = 0.7
		tmp = np.sum(x*w)+b
		if tmp<=0:
				return 0
		else :
				return 1

def perceptron_OR(x1,x2):
		x = np.array([x1,x2])
		w = np.array([0.5,0.5])
		b = -0.2
		tmp = np.sum(x*w)+b
		if tmp<=0:
				return 0
		else :
				return 1
```

perceptron 한계: XOR 게이트 선형적 분리가 불가능하다.

- MLP(multy layer percepton)로 비선형 분리 문제 해결
- 층을 깊게 쌓는다 == 다양한 표현이 가능해진다.

```python
def XOR(x1,x2):
  x3 = perceptron_OR(x1,x2)
  x4 = perceptron_NAND(x1,x2)
  y  = perceptron_AND(x3,x4)
  return y 
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```
