# 2. EDA

- ID : 집을 구분하는 번호
- date : 집을 구매한 날짜
- price : 집의 가격(Target variable)
- bedrooms : 침실의 수
- bathrooms : 화장실의 수
- sqft_living : 주거 공간의 평방 피트(면적)
- sqft_lot : 부지의 평방 피트(면적)
- floors : 집의 층 수
- waterfront : 집의 전방에 강이 흐르는지 유무 (a.k.a. 리버뷰)
- view : 집이 얼마나 좋아 보이는지의 정도
- condition : 집의 전반적인 상태
- grade : King County grading 시스템 기준으로 매긴 집의 등급
- sqft_above : 지하실을 제외한 평방 피트(면적)
- sqft_basement : 지하실의 평방 피트(면적)
- yr_built : 지어진 년도
- yr_renovated : 집을 재건축한 년도
- zipcode : 우편번호
- lat : 위도
- long : 경도
- sqft_living15 : 2015년 기준 주거 공간의 평방 피트(면적, 집을 재건축했다면, 변화가 있을 수 있음)

# 3.Base Model

```python
train = pd.read_csv(url+'/train.csv')
test  = pd.read_csv(url+'/test.csv')
target = train.price
```

```python
df = pd.concat([train,test],ignore_index=True)
df['date'] = df['date'].apply(lambda x : str(x[:6])).astype(int)
df.drop(['id','date','price'],axis=1,inplace=True)
train = df.iloc[:len(train),:]
test = df.iloc[len(train):,:]
```

```python
skew_columns = ['bedrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement']

for c in skew_columns:
    df[c] = np.log1p(df[c].values)
```

```python
x_train,x_test, y_train,y_test = train_test_split(train,target,random_state=42)
```

```python
rf  = RandomForestRegressor()
xgb = XGBRegressor()
lgbm = LGBMRegressor()
models =[rf,xgb,lgbm]
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt


def rmse(y_actual,y_pred):
    return sqrt(mean_squared_error(y_actual, y_pred))
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(rmse(y_test,y_pred))
```

```python
blend= rf.predict(x_test)*0.05+lgbm.predict(x_test)*0.05+xgb.predict(x_test)*0.9
rmse(y_test,blend)
```

- 아무런 조정없이 단순 id와 date 제거만으로 xgb의 성능이 약 13000의 에러값을 보인다.
- 블랜드의 결과는 더욱 안좋다

# 3.1 Base model up

```python
train = pd.read_csv(url+'/train.csv')
test  = pd.read_csv(url+'/test.csv')
target = np.log1p(train.price)

df = pd.concat([train,test],ignore_index=True)
df['date'] = df['date'].apply(lambda x : str(x[:6])).astype(int)
df.drop(['id','date','price'],axis=1,inplace=True)

skew_columns = ['bedrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement']
for c in skew_columns:
    df[c] = np.log1p(df[c].values)
    
train = df.iloc[:len(train),:]
test = df.iloc[len(train):,:]

x_train,x_test, y_train,y_test = train_test_split(train,target,random_state=42)
```

```python
rf  = RandomForestRegressor()
xgb = XGBRegressor()
lgbm = LGBMRegressor()
models =[rf,xgb,lgbm]
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(rmse(np.expm1(y_test),np.expm1(y_pred)))
```

- 소수의 변수 제거, 소수의 변수 log 변환 그리고 target값의 log변환으로 충분히 좋은 성능을 끌어냈다.

# 4. FE 

```python
train = pd.read_csv(url+'/train.csv')
test  = pd.read_csv(url+'/test.csv')
target = np.log1p(train.price)
df = pd.concat([train,test],ignore_index=True)
```

```python
df['real_quft'] =  df.sqft_lot - df.sqft_living
df['real_quft']+=df['real_quft'].min()
df['real_built'] = df.apply(lambda row: row.yr_renovated if row.yr_renovated != 0 else row.yr_built,axis=1)
df['total_rooms'] = df['bedrooms']+df['bathrooms']
df['total_grade'] = df.grade +df.condition+df.view
```

```python
df.drop(['id','date','lat','long','price'],axis=1,inplace=True)
```

```python
skew_feature = [ 'sqft_lot', 'waterfront', 'sqft_lot15', 'yr_renovated', 'view']
```

```python
for i in skew_feature:
    df[i] = np.log1p(df[i])
```

```python
zip =pd.get_dummies(df.zipcode)
df=pd.concat([df,zip],axis=1)
```

```python
rf = RandomForestRegressor(n_estimators=1000)
xgb = XGBRegressor(n_estimators=1000)
lgbm=LGBMRegressor(n_estimators=1100,max_depth=30)
```

```python
train = df.iloc[:len(train)]
test = df.iloc[len(train):]
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,target,random_state=42)
```

```python
models = [rf,lgbm,xgb]
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(rmse(np.expm1(y_test),np.expm1(y_pred)))
    
```

```python
blend= rf.predict(x_test)*0.1+lgbm.predict(x_test)*0.1+xgb.predict(x_test)*0.8
rmse(np.expm1(y_test),np.expm1(blend))
```

- 파생변수 생성과 ,log 정규화로 예측값이 좋아진다

# 2차 개선

```python
train = pd.read_csv(url+'/train.csv')
test  = pd.read_csv(url+'/test.csv')
target = np.log1p(train.price)
df = pd.concat([train,test],ignore_index=True)
```

```python
df.drop(['id','date','price'],axis=1,inplace=True)

df.long = abs(df.long)

df['real_built'] = df.apply(lambda row: row.yr_renovated if row.yr_renovated != 0 else row.yr_built,axis=1)
df['total_rooms'] = df['bedrooms']+df['bathrooms']
df['real_quft'] =  df.sqft_lot - df.sqft_living
df['real_quft']=abs(df['real_quft'])
df['total_grade'] = df.grade +df.condition+df.view

df['yr_built'] = df['yr_built']//10
df['yr_renovated'] = df['yr_renovated']//10
df['real_built'] = (df['real_built']//10).astype('int')

for i in df.columns:          # long -값이기에 log취하면 nan뜬다 ->abs취하거나 드랍
    df[i] = np.log1p(df[i])
    
zip =pd.get_dummies(df.zipcode)
df=pd.concat([df,zip],axis=1)
```

```python
train = df.iloc[:len(train)]
test = df.iloc[len(train):]

xgb = XGBRegressor()
lgbm=LGBMRegressor(n_estimators=1100,max_depth=30)
rf  = RandomForestRegressor()
models = [rf,lgbm,xgb]

x_train,x_test,y_train,y_test = train_test_split(train,target,random_state=42)
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(rmse(np.expm1(y_test),np.expm1(y_pred)))
```

```python
blend= rf.predict(x_test)*0.1+lgbm.predict(x_test)*0.55+xgb.predict(x_test)*0.35
rmse(np.expm1(y_test),np.expm1(blend))
```

```python
lgbm = LGBMRegressor(n_estimators=1100,max_depth=30)
```

# 회고

- 전 노드인 bike sharing demand에서와 비슷하게 최대한 많은 EDA와 FE를 시도해 보았다. 하지만 성능이 꾸준히 개선은 되지만 가장 기본적인 base 모델에 몇가지 FE만 한 모델의 성능이 가장 좋게나오는 역설적 상황이 나왔다. 

- 이를 더 공부해 보아야겠지만, 아무래도 나의 분석 순서 또는 FE이 불필요한 부분이 많다고 생각된다. 또한 쉘의 시행 순서가 뒤죽박죽이어서 쉘을 재실행 하였을시 오류가 빈번히 발생하였다. 이러한 악습관을 고쳐야겠다. 이번 노드를 하면서 느낀점은 EDA를 통하여 데이터를 파악하는것은 좋아도 FE은 최대한 단순화하고 가볍게 변형을 가하여 모델의 성능을 높여야 한다는 점이다. 이를 잘하기 위하여 EDA ,FE ,model에 대한 전반적인 공부를 다시해야겠다고 생각하였다.

- 또한 캐글의 점수가 부족할떄 train데이터를 통째로 학습하면 성능이 더욱 개선된다.
