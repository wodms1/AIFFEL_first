```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.metrics import classification_report,confusion_matrix,plot_confusion_matrix
```

```python
data = load_digits()
```

# 3.Model

```python
x_train ,x_test ,y_train , y_test = train_test_split(digit_data ,digit_label, random_state=42, test_size =0.2)
    # train,test data로 분리 train : test = 4 : 1
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
svm_model = svm.SVC()
sgd = SGDClassifier()
lr = LogisticRegression()

model =[dt,rf,svm_model,sgd,lr]

# hyper - prameter 은 default 값으로 설정
```

```python
for clf in model:
  clf.fit(x_train,y_train)
  y_pred = clf.predict(x_test)
  print(str(clf))
  print(classification_report(y_test,y_pred))  
  print(confusion_matrix(y_test,y_pred))       
  plot = plot_confusion_matrix(clf,x_test,y_test)
  plt.show()

  print('----------')
```

-  다중 분류 문제이면서 각 target별 data가 작기에 accuracy보다는 조화평균인 f1 socre로 모델을 평가한다.
- 모델들 중 ensemble 모델이 가장 효율적이라 생각하였으나 실질적으로 SVM모델이 accuracy 0.99와 높은 f1-score를 기록하여 svm이 생각보다 더 좋은 모델이라는 생각이 들었다.
