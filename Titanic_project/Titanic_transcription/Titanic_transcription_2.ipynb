# EDA

## taget data check

```python
f,ax = plt.subplots(1,2,figsize=(18,8))# subplots ->s붙여야
data.Survived.value_counts().plot.pie(explode=[0,0.1],
                                      autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Survived')
ax[0].set_ylabel('')
sns.countplot('Survived',data=data,ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```

- 38.4%민 생존
- 순서형 feature -> PClass
- 연속형 -> Age

## Sex ->Categorical Feature

```python
data.groupby(["Sex",'Survived'])['Survived'].count()
# "Sex",'Survived'을 묶고 이를 생존으로 카운팅
```

```python
f , ax = plt.subplots(1,2,figsize=(18,8))
data[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex',hue='Survived',data = data,ax=ax[1])
ax[1].set_title('Sex:Sirvived vs Dead')
plt.show()
```

- 남성이 여성에비해 배에 많이 탑승하였다.
- 여성이 남성에 비해 많이 생존하였다.

## Pclass ->Ordinal Feature

```python
pd.crosstab(index=data.Pclass,columns=data.Survived,margins=True).style.background_gradient(cmap='summer_r')
    # ->.style.background_gradient(cmap='색상')
```

```python
f,ax = plt.subplots(1,2,figsize=(18,8))
data['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Numner Of Passengers By Pclass')
ax[0].set_ylabel('Count')
sns.countplot('Pclass',hue='Survived',data=data,ax=ax[1])
ax[1].set_title('Pclass:Survived vs Dead')
plt.show()
```
